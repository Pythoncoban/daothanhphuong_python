### Thực hành: Thêm học sinh vào lớp

```python
list_student = ['Andy', 'Bryan', 'Cindy', 'Dan']
new_student = ['Emma', 'Ferb']
list_student.extend(new_student)
print(list_student)
```

### Kiểm tra phần tử tồn tại trong list

```python
list_student = ['Andy', 'Bryan', 'Cindy', 'Dan']
new_student = ['Emma', 'Ferb']
print('Steve' in list_student)
```

### Xóa phần tử khỏi list

```python
list_student = ['Andy', 'Bryan', 'Cindy', 'Dan']
new_student = ['Emma', 'Ferb']
# remove
list_student.remove('Andy')
print(list_student)
```

```python
list_student = ['Andy', 'Bryan', 'Cindy', 'Dan']
new_student = ['Emma', 'Ferb']
# del -> Xóa phần tử tại vị trí tại index đã được chỉ định trước
del list_student[0]
print(list_student)
```

### Sắp xếp list

```python
list_student = ['Andy', 'Bryan', 'Kindy', 'Dan']
new_student = ['Emma', 'Ferb']
# list.sort --> sắp xếp mặc định từ thấp tới cao
list_student.sort()
print(list_student)

# list.sort(rervert = True) --> Sắp xếp cao --> thấp
list_student.sort(reverse = True)
print(list_student)
```
