```python
class Net:
    def __init__(self, province, nodeB, cable, mec, pole, protection, area):
        self.province = province
        self.nodeB = nodeB
        self.cable = cable
        self.mec = mec
        self.pole = pole
        self.protection = protection
        self.area = area
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import csv

# tạo cửa sổ ứng dụng, Treeview
window = Tk()
wd = Tk()

# Cấu hình Treeview
col = ('province', 'nodeB', 'cable', 'mec', 'pole', 'protection', 'area')
tr = ttk.Treeview(wd, columns = col, show = 'headings', heigh = 10)
style = ttk.Style(wd)
ttk.Style().configure(".", font=('Helvetica', 8), foreground="Red")

tr.column('province', width = 100)
tr.column('nodeB', width = 100)
tr.column('cable', width = 100)
tr.column('mec', width = 100)
tr.column('pole', width = 100)
tr.column('protection', width = 100)
tr.column('area', width = 100)

tr.heading('province', text = 'province')
tr.heading('nodeB', text = 'nodeB')
tr.heading('cable', text = 'cable')
tr.heading('mec', text = 'mec')
tr.heading('pole', text = 'pole')
tr.heading('protection', text = 'protection')
tr.heading('area', text = 'area')

# setup thông tin cho window
window.title("Network Management System")
window.geometry("700x250")

# setup thông tin cho window Treeview
wd.title("Network Management System")
wd.geometry("700x200")

# tạo list để lưu trữ toàn bộ các object
list_Net = []


# mô tả các function cho ứng dụng
# add/search/update/delete/reset

def add_Net():
    
    # 1. Lấy dữ liệu từ Entry
    
    province_value = str(province.get())          
    nodeB_value = int(nodeB.get())
    cable_value = int(cable.get())
    mec_value = int(mec.get())
    pole_value = int(pole.get())
    protection_value = str(protection.get())
    area_value = str(area.get())
    
    # 2. Kiểm tra province input đã tồn tại chưa và báo lỗi nhập trùng 
    
    add_result = True
    for i in list_Net:
        if i.province == province_value:
            messagebox.showinfo("Add Net", f"{province_value} already exist!!!")
            add_result = False
            break
            
    # 3. Tiến hành Add nếu dữ liệu nhập chuẩn & không trùng province   
    
    if add_result == True:
        # 3.1 Tạo đối tượng class Net đệm để add vào list_Net
        new_Net = Net(province_value, nodeB_value, cable_value, mec_value, pole_value, protection_value, area_value)
        list_Net.append(new_Net)

        # 3.2 Add Treeview
        tr.insert('', END, values=(province_value, nodeB_value, cable_value, mec_value, pole_value, protection_value, area_value))
        
        # 3.3 Hiển thị thông báo Add thành công
        messagebox.showinfo("Add Net", f"{province_value} is added successfully!") 
        
    
        # 3.4 Xóa giá trị ở các entry
        province.delete(0, END)
        nodeB.delete(0, END)
        cable.delete(0, END)
        mec.delete(0, END)
        pole.delete(0, END)
        protection.delete(0, END)
        area.delete(0, END)
        
    # 4. Hiển thị list_Net.provine sau khi add
    print([item.province for item in list_Net]) 


def import_new_Net():
    with open('D:\Import_New.csv') as f:
        reader = csv.reader(f)
        list_new = [row for row in reader]
        
    # check dữ liệu nhập vào có bị lặp dữ liệu province không
    import_new_result = True
    for i in range(1, len(list_new)-1):
        for j in range(i+1,len(list_new)):
            if list_new[i][0] == list_new[j][0]:
                messagebox.showinfo("Wrong file", f"province {list_new[i][0]} appear more than 1 time")
                import_new_result = False
                break
    
    # check dữ liệu nhập vào có bị trùng dữ liệu province trong database hiện tại không
    if import_new_result == True:
    
        # Tạo list_temp_new dữ liệu province của List_Net
        list_temp_new = []
        for i in range(len(list_Net)):
                list_temp_new.append(list_Net[i].province)
        for i in range(1, len(list_new)):
                if list_new[i][0] in list_temp_new:
                    messagebox.showinfo("Wrong file", f"province {list_new[i][0]} has been already existed in database.\nPlease check input file again!!!")
                    import_new_result = False
                    break 
    
    # Tạo mới dữ liệu sau khi đã kiểm tra          
    if import_new_result == True:
        for i in range(1, len(list_new)):
            province_value = list_new[i][0]
            nodeB_value = list_new[i][1]
            cable_value = list_new[i][2]
            mec_value = list_new[i][3]
            pole_value = list_new[i][4]
            protection_value = list_new[i][5]
            area_value = list_new[i][6]
    
            new_Net = Net(province_value, nodeB_value, cable_value, mec_value, pole_value, protection_value, area_value)
        
            list_Net.append(new_Net)

            # Add Treeview
            tr.insert('', END, values=(province_value, nodeB_value, cable_value, mec_value, pole_value, protection_value, area_value))
            
        # hiển thị thông báo
        messagebox.showinfo("Add Net", f"Import_New is successfully!")

        # Hiển thị list_Net.provine sau khi add
        print([item.province for item in list_Net])
        

def search_Net():
    search_province_value = str(search_province.get())
    search_result = False
    for i in list_Net:
        if i.province == search_province_value:
            search_result = True
            messagebox.showinfo("Search result: ", f"province: {i.province} \nNodeB: {i.nodeB} \nCable: {i.cable} \nMEC: {i.mec} \nPole: {i.pole} \nProtection: {i.protection}\nArea: {i.area}")
            break
    if search_result == False:
        messagebox.showinfo("Search Net", f"{search_province_value} not found!")
        
def update_Net():
    # lấy giá trị province từ các widget
    province_value = str(province.get()) if province.get() != "" else ""          
    nodeB_value = int(nodeB.get()) if nodeB.get() != "" else 0
    cable_value = int(cable.get()) if cable.get() != "" else 0
    mec_value = int(mec.get()) if mec.get() != "" else 0  
    pole_value = int(pole.get()) if pole.get() != "" else 0  
    protection_value =  str(protection.get()) if protection.get() != "" else ""
    area_value =  str(area.get()) if area.get() != "" else ""

    # Kiểm tra và update thông tin
    update_result = False
    for i in range(len(list_Net)):
        if list_Net[i].province == province_value:
            update_result = True
            result= messagebox.askquestion('Update_Net',f'Do you want to update {list_Net[i].province} ?')
            if result=='yes':
                if nodeB_value >0:
                    list_Net[i].nodeB = nodeB_value
                if cable_value >0:
                    list_Net[i].cable = cable_value
                if mec_value >0:
                    list_Net[i].mec = mec_value
                if pole_value >0: 
                    list_Net[i].pole = pole_value
                if protection_value != "":
                    list_Net[i].protection = protection_value
                if area_value != "":
                    list_Net[i].area = area_value
                break
            
    # Update Treeview (add mới và del cũ)
    tr.insert('', i+1, values=(list_Net[i].province, list_Net[i].nodeB, list_Net[i].cable, list_Net[i].mec, list_Net[i].pole, list_Net[i].protection, list_Net[i].area))
    item = tr.get_children()[i]
    tr.delete(item)
    

    # Thông báo update thành công
    if update_result == True and result=='yes':
        messagebox.showinfo("Update Net", f"{province_value} is updated successfully")
        
        # xóa giá trị ở các entry
        province.delete(0, END)
        nodeB.delete(0, END)
        cable.delete(0, END)
        mec.delete(0, END)
        pole.delete(0, END)
        protection.delete(0, END)
        area.delete(0, END)
        
    
    elif  update_result == False:
        messagebox.showinfo("Update Net", f"{province_value} is updated False")
        
def import_update_Net():
    with open('D:\Import_Update.csv') as f:
        reader = csv.reader(f)
        list_update = [row for row in reader]
    
    # check dữ liệu nhập vào 
    import_update_result = True
    
    # check dữ liệu nhập vào có bị lặp dữ liệu province không 
    for i in range(1, len(list_update)-1):
        for j in range(i+1,len(list_update)):
            if list_update[i][0] == list_update[j][0]:
                messagebox.showinfo("Wrong file", f"province {list_update[i][0]} appear more than 1 time")
                import_update_result = False
                break
    
    # check dữ liệu nhập vào đã có trong database chưa
    if import_update_result == True:
        # Tạo list_temp tên các province của List_Net
        list_temp_update = []
        for i in range(len(list_Net)):
            list_temp_update.append(list_Net[i].province)
        
            
        for i in range(1, len(list_update)):
                if list_update[i][0] not in list_temp_update:
                    import_update_result = False
                    messagebox.showinfo("Wrong file", f"province {list_update[i][0]} not found in database.\nRe-correct the update file again!!!")
                    break

    # Cập nhật dữ liệu sau khi đã kiểm tra đầu vào
    # Hỏi người dùng có chắc chắn muốn cập nhật không?
    result= messagebox.askquestion('Import update',f'Do you want to Import update?')
    if result=='yes':
        if import_update_result == True:
            for i in range(1, len(list_update)):
                for j in range(len(list_Net)):
                    if list_update[i][0] == list_Net[j].province:
                        list_Net[j].nodeB = list_update[i][1]
                        list_Net[j].cable = list_update[i][2]
                        list_Net[j].MEC = list_update[i][3]
                        list_Net[j].pole = list_update[i][4]
                        list_Net[j].protection = list_update[i][5]
                        list_Net[j].area = list_update[i][6]
                   
            # Del Treeview cũ và làm mới Treeview
            delete_tree()
            treeview_Net()
            
            # hiển thị thông báo
            messagebox.showinfo("Improt update Net", f"Import update is successfully!")
            
            
    
def delete_Net():
    search_province_value = search_province.get()
    
    del_result = False
    
    for i in range(len(list_Net)):
        if list_Net[i].province == search_province_value:
            del_result = True
            result= messagebox.askquestion('Delete_Net',f'Do you want to del {list_Net[i].province} ?')
            if result=='yes':
                del list_Net[i]
                
                # Xóa trên Treeview
                item = tr.get_children()[i]
                tr.delete(item)
                
                # Thông báo xóa thành công
                messagebox.showinfo("Delete Net", f"Delete {search_province_value} successfully!")
                
                break
            
    # Thông báo không tìm thấy đối tượng để xóa
    if del_result == False:
        messagebox.showinfo("Delete Net", "Province not found!")
    
    
    print([item.province for item in list_Net])


def reset():
    province.delete(0, END)
    nodeB.delete(0, END)
    cable.delete(0, END)
    mec.delete(0, END)
    pole.delete(0, END)
    protection.delete(0, END)
    area.delete(0, END)
    search_province.delete(0, END)

def export_Net():
    with open('D:\Export.csv','w') as f:
        writer = csv.writer(f)
        writer.writerow(['province', 'nodeB', 'cable', 'mec', 'pole', 'protection', 'area'])
        
        for i in range(len(list_Net)):
            list_export_temp = []
            list_export_temp.append(list_Net[i].province)
            list_export_temp.append(list_Net[i].nodeB)
            list_export_temp.append(list_Net[i].cable)
            list_export_temp.append(list_Net[i].mec)
            list_export_temp.append(list_Net[i].pole)
            list_export_temp.append(list_Net[i].protection)
            list_export_temp.append(list_Net[i].area)
            writer.writerow(list_export_temp)
   
    messagebox.showinfo("Export Net", f"Export Net successfully!")
    
def report_Net():
    with open('D:\Report.csv','w') as f:
        writer = csv.writer(f)
        writer.writerow(['area', 'nodeB', 'cable', 'pole'])
        # Tạo list các area (sau khi loại bỏ trùng)
        list_area = []
        for i in range(len(list_Net)):
            list_area.append(list_Net[i].area)
        s= set(list_area)
        list_area = list(s)

        # Tổng hợp các thông tin NodeB, cable, pole theo area
        for i in range(len(list_area)):
            list_report_temp = []
            sum_nodeB = 0
            sum_cable = 0
            sum_pole = 0
            for j in range(len(list_Net)):
                if list_area[i] == list_Net[j].area:
                    sum_nodeB += int(list_Net[j].nodeB)
                    sum_cable += int(list_Net[j].cable)
                    sum_pole += int(list_Net[j].pole)
            list_report_temp.append(list_area[i])
            list_report_temp.append(sum_nodeB)
            list_report_temp.append(sum_cable)
            list_report_temp.append(sum_pole)
            writer.writerow(list_report_temp)
   
    messagebox.showinfo("Report Net", f"Export Net successfully!")


def treeview_Net():
    
    data = []
    for i in range(len(list_Net)):
        data.append((list_Net[i].province, list_Net[i].nodeB, list_Net[i].cable, list_Net[i].mec, list_Net[i].pole, list_Net[i].protection, list_Net[i].area ))
    for data in data:
        tr.insert('', END, values=data)

def delete_tree():
    for item in tr.get_children():
        tr.delete(item)





# setup các widget
# tạo label
province_lbl = Label(window, text="Province")
NodeB_lbl = Label(window, text="NodeB")
cable_lbl = Label(window, text="Cable")
mec_lbl = Label(window, text="MEC")
pole_lbl = Label(window, text="Pole")
protection_lbl = Label(window, text="Protection")
area_lbl = Label(window, text="Area")

# tạo entry
province = Entry(window)
nodeB = Entry(window)
cable = Entry(window)
mec = Entry(window)
pole = Entry(window)
protection = Entry(window)
area = Entry(window)
search_province = Entry(window)



# tạo button
add_btn = Button(window, text="Add", command=add_Net)
import_new_btn = Button(window, text="Import New", command=import_new_Net)
import_update_btn = Button(window, text="Import Update", command=import_update_Net)
search_btn = Button(window, text="Search", command=search_Net)
update_btn = Button(window, text="Update", command=update_Net)
delete_btn = Button(window, text="Delete", command=delete_Net)
reset_btn = Button(window, text="Reset", command=reset)
export_btn = Button(window, text="Export", command=export_Net)
report_btn = Button(window, text="Report Net", command=report_Net)

# dán các widget lên window
province_lbl.grid(row=0, column=0)
NodeB_lbl.grid(row=1, column=0)
cable_lbl.grid(row=2, column=0)
mec_lbl.grid(row=0, column=2)
pole_lbl.grid(row=1, column=2)
protection_lbl.grid(row=2, column=2)
area_lbl.grid(row=0, column=4)
province.grid(row=0,column=1)
nodeB.grid(row=1,column=1)
cable.grid(row=2, column=1)
mec.grid(row=0, column=3)
pole.grid(row=1, column=3)
protection.grid(row=2, column=3)
area.grid(row=0, column=5)
search_province.grid(row=3,column=0)
search_btn.grid(row=3,column=1)
delete_btn.grid(row=3,column=2)
add_btn.grid(row=3,column=3)
update_btn.grid(row=3, column=4)
reset_btn.grid(row=4,column=1)
import_new_btn.grid(row=4,column=3)
import_update_btn.grid(row=4, column=4)
export_btn.grid(row=4, column=2)
report_btn.grid(row=3, column=5)
tr.grid(row=0, column=0)


# chạy ứng dụng
window.mainloop()
```
